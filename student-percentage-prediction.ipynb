**Student Percentage Prediction Model (Linear Regression)**

1. Importing the required libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

2. Reading the given data

```python
link = "http://bit.ly/w-data"
data = pd.read_csv(link)

print("The student data has been successfully imported and its first ten entries look like the following:")
data.head(10)
```

3. Plotting the imported data

```python
data.plot(x = "Hours", y = "Scores", style = "*")
plt.title("Hours vs Percentage")
plt.xlabel("Hours Studied")
plt.ylabel("Score Obtained")
plt.show()
```

4. Splitting the data into training and test sets

```python
X = data.iloc[:, :-1].values
y = data.iloc[:, 1].values

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

5. Training the algorithm

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)

print("Training complete.")
```

6. Plotting the regression line

```python
line = regressor.coef_*X+regressor.intercept_

plt.scatter(X, y)
plt.plot(X, line)
plt.show()
```

7. Making predictions and comparing them with actual values

```python
print(X_test)
y_pred = regressor.predict(X_test)

df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
df

hours = 9.25
input = np.array([hours])
input = input.reshape(-1, 1)
own_pred = regressor.predict(input)
print("No. of hours = {}".format(hours))
print("Predicted score = {}".format(own_pred[0]))
```

8. Evaluating the model

```python
from sklearn import metrics
print("Mean Absolute Error:", metrics.mean_absolute_error(y_test, y_pred))
```
